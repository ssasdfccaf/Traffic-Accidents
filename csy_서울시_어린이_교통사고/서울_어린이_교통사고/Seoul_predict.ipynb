# 예측 vs 실제

# 실제

1 양천구 / 2 강남구 / 3 송파구 

```python
# 실제
# 1=양천구 ,2=강남구, 3=송파구

import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = [1, 2, 3]
y = [4, 5, 6]

fig, ax = plt.subplots()

bars = ax.bar(x, y, color='lightgray', edgecolor='black')
bars[0].set_hatch('x')
bars[1].set_hatch('O')
bars[2].set_hatch('.')

plt.tight_layout()
plt.show()
```

# 예측 1

어린이 사고 - 예측 

```python
# 예측 = yello

# 1=은평구 ,2=강남구, 3=강동구

import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

fig, ax = plt.subplots()

# 진한 노란색
bars = ax.bar(x, y, color='#FFD700', edgecolor='black')  
bars[0].set_hatch('x')
bars[1].set_hatch('O')
bars[2].set_hatch('.')

plt.tight_layout()
plt.show()
```
