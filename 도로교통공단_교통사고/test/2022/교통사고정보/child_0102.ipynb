```python
# 각 연령대의 치사율 비교
import matplotlib.pyplot as plt
import pandas as pd

age_groups = ["20세이하", "21-30세", "31-40세", "41-50세", "51-60세", "61-64세", "65세이상", "불명"]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

# 치사율 계산
mortality_rate = [death / accident * 100 if accident > 0 else 0 for death, accident in zip(death_counts, accident_counts)]

df = pd.DataFrame({
    '연령대': age_groups,
    '치사율': mortality_rate
})

# 점 그래프
plt.figure(figsize=(10, 6))
plt.scatter(df['연령대'], df['치사율'], s=df['치사율']*10, alpha=0.5, color='coral')
plt.title('가해운전자 연령대별 교통사고 치사율')
plt.xlabel('연령대')
plt.ylabel('치사율 (%)')
plt.show()

# 원형 그래프
plt.figure(figsize=(8, 8))
plt.pie(df['치사율'], labels=df['연령대'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('가해운전자 연령대별 교통사고 치사율')
plt.show()

# 막대 그래프
plt.figure(figsize=(10, 6))
plt.bar(df['연령대'], df['치사율'], color='skyblue')
plt.title('가해운전자 연령대별 교통사고 치사율')
plt.xlabel('연령대')
plt.ylabel('치사율 (%)')
plt.show()

# 선 그래프
plt.figure(figsize=(10, 6))
plt.plot(df['연령대'], df['치사율'], marker='o', color='orange')
plt.title('가해운전자 연령대별 교통사고 치사율')
plt.xlabel('연령대')
plt.ylabel('치사율 (%)')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 정의
age_groups = ["20세이하", "21-30세", "31-40세", "41-50세", "51-60세", "61-64세", "65세이상", "불명"]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

# 치사율 계산
mortality_rate = [death / accident * 100 if accident > 0 else 0 for death, accident in zip(death_counts, accident_counts)]

# 데이터 프레임 생성
df = pd.DataFrame({
    '연령대': age_groups,
    '치사율': mortality_rate
})

# 원형 그래프
plt.figure(figsize=(8, 8))
plt.pie(df['치사율'], labels=df['연령대'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('가해운전자 연령대별 교통사고 치사율')
plt.show()
```

```python
# 히트맵 - 치사율 중심

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

age_groups = ["20세이하", "21-30세", "31-40세", "41-50세", "51-60세", "61-64세", "65세이상", "불명"]
accident_counts = [6508, 26904, 28402, 33915, 44581, 19160, 34652, 2714]
death_counts = [72, 335, 317, 409, 570, 297, 735, 0]
injury_counts = [8863, 39706, 41287, 48389, 63632, 27566, 49281, 3079]

# 데이터 프레임 생성
df = pd.DataFrame({
    '연령대': age_groups,
    '사고건수': accident_counts,
    '사망자수': death_counts,
    '부상자수': injury_counts
})

plt.figure(figsize=(10, 6))
sns.heatmap(df.set_index('연령대').T, annot=True, cmap='YlGnBu', fmt='d', linewidths=.5, cbar=True)
plt.title('가해운전자 연령대별 교통사고 현황')
plt.show()
```
