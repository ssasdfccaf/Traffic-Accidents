```python
# 극좌표계 레이더 차트
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc 
rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False  

# 데이터 정의
months = ["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]
accidents = [2668, 2104, 2350, 2994, 3291, 2994, 3118, 3100, 3403, 3658, 3346, 2888]

# 그래프 생성
plt.figure(figsize=(10, 10))

# 각 월에 대한 각도 계산
theta = np.linspace(0, 2*np.pi, len(months), endpoint=False)

# 극좌표계 레이더 차트 생성
plt.polar(theta, accidents, marker='o', linestyle='-')

# 각도에 따른 레이블 설정
plt.xticks(theta, months)

# 그래프 설정
plt.title('극좌표계 레이더 차트 - 노인 교통사고 월별 사고건수')
plt.show()
```

```python
# Density Plot 분포도 시각화
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 정의
months = ["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]
accidents = [2668, 2104, 2350, 2994, 3291, 2994, 3118, 3100, 3403, 3658, 3346, 2888]

# 그래프 생성
plt.figure(figsize=(10, 6))

# 밀도 플로트 생성
sns.kdeplot(data=accidents, shade=True)

# 그래프 설정
plt.title('노인 교통사고 월별 사고건수 분포')
plt.xlabel('사고건수')
plt.xticks(range(0, max(accidents)+500, 500))

# 그래프 표시
plt.show()
```

```python
# 월별
import matplotlib.pyplot as plt

# 데이터 정의
months = ["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]
accidents = [2668, 2104, 2350, 2994, 3291, 2994, 3118, 3100, 3403, 3658, 3346, 2888]
fatalities = [78, 77, 69, 111, 98, 101, 110, 121, 132, 131, 128, 102]
injuries = [2863, 2267, 2540, 3310, 3623, 3272, 3416, 3356, 3697, 4067, 3612, 3169]

# 그래프 생성
plt.figure(figsize=(10, 6))

# 선 그래프 생성
plt.plot(months, accidents, label='사고건수', marker='o')
plt.plot(months, fatalities, label='사망자수', marker='o')
plt.plot(months, injuries, label='부상자수', marker='o')

# 그래프 제목 및 축 라벨 설정
plt.title('노인 교통사고 월별 통계')
plt.xlabel('월')
plt.ylabel('건수')
plt.legend()  # 범례 표시

# 그래프 표시
plt.show()
```
