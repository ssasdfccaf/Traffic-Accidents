```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc 
rc('font', family='AppleGothic') 		
plt.rcParams['axes.unicode_minus'] = False  


# 주어진 데이터를 DataFrame으로 만들기
data = {
    '시도': ['서울', '서울', '서울', '서울', '부산', '부산', '경기', '경기', '경기', '경기'],
    '시군구': ['종로구', '중구', '용산구', '성동구', '중구', '서구', '수원시', '성남시', '의정부시', '안양시'],
    '발생건수': [191, 164, 170, 158, 73, 105, 666, 522, 264, 268]
}

df = pd.DataFrame(data)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(df['시군구'], df['발생건수'], color='skyblue')
plt.title('시도, 시군구별 교통사고 발생건수')
plt.xlabel('시군구')
plt.ylabel('발생건수')
plt.show()
```
