```python
from bs4 import BeautifulSoup

html_code = """
<!DOCTYPE html PUBLIC "-//W3C//DTD HTML 4.01 Transitional//EN" "http://www.w3.org/TR/html4/loose.dtd">
<html>
<head>
<meta http-equiv="Content-Type" content="text/html; charset=UTF-8">
<title>Insert title here</title>
</head>
<body>

<style type="text/css">
table {width:100%;border-top:1px solid #737678;}
td {border:1px solid #cccccc;text-align:center;vertical-align:middle;}
</style>

<h2>보행자 교통사고</h2>
<table>
    <thead>
        <tr bgcolor="#ebf2f7">
            <th>년도</th>
            <th>사고건수(건)</th> 
            <th>사망자수(명)</th>
            <th>부상자수(명)</th>
        </tr>
    </thead>
    <tbody>
        
        <tr>
            <td>2018</td>
            <td>45,921</td>
            <td>1,487</td>
            <td>46,400</td>
        </tr>
        
        <tr>
            <td>2019</td>
            <td>46,682</td>
            <td>1,302</td>
            <td>47,200</td>
        </tr>
        
        <tr>
            <td>2020</td>
            <td>36,601</td>
            <td>1,093</td>
            <td>36,939</td>
        </tr>
        
        <tr>
            <td>2021</td>
            <td>35,665</td>
            <td>1,018</td>
            <td>36,001</td>
        </tr>
        
        <tr>
            <td>2022</td>
            <td>37,611</td>
            <td>933</td>
            <td>38,088</td>
        </tr>
        
    </tbody>
</table>

</body>
</html>
"""

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_code, 'html.parser')

# 테이블 데이터 추출
table = soup.find('table')
rows = table.find_all('tr')

# 각 행의 데이터를 출력
for row in rows[1:]:  # 첫 번째 행은 헤더이므로 무시하고 나머지 행부터 출력
    columns = row.find_all('td')
    year = columns[0].get_text()
    accidents = columns[1].get_text()
    deaths = columns[2].get_text()
    injuries = columns[3].get_text()
    
    print(f"Year: {year}, Accidents: {accidents}, Deaths: {deaths}, Injuries: {injuries}")
```

```python
import matplotlib.pyplot as plt

# 데이터
years = ['2018', '2019', '2020', '2021', '2022']
accidents = [45921, 46682, 36601, 35665, 37611]
deaths = [1487, 1302, 1093, 1018, 933]
injuries = [46400, 47200, 36939, 36001, 38088]

# 선 그래프 그리기
plt.figure(figsize=(10, 6))

# 사고건수
plt.plot(years, accidents, marker='o', label='Accidents', color='blue')

# 사망자수
plt.plot(years, deaths, marker='o', label='Deaths', color='red')

# 부상자수
plt.plot(years, injuries, marker='o', label='Injuries', color='green')

# 그래프 제목과 라벨 설정
plt.title('Pedestrian Traffic Accidents (2018-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')

# 범례 추가
plt.legend()

# 그래프 표시
plt.grid(True)
plt.show()
```
