```python
# 각 행정구역의 사고건수 비교 

import pandas as pd
import matplotlib.pyplot as plt
from math import pi

# 주어진 데이터를 DataFrame으로 변환
data = {
    "행정구역별(1)": ["전국", "서울특별시", "부산광역시"],
    "사고건수 (건)": [8889, 1026, 443]
}

df = pd.DataFrame(data)

# 각 변수의 각도 계산
categories = list(df['행정구역별(1)'])
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# 그래프 그리기
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories, color='grey', size=10)
ax.set_rlabel_position(0)
plt.yticks([2000, 4000, 6000, 8000], ["2000", "4000", "6000", "8000"], color="grey", size=8)
plt.ylim(0, 9000)

# 데이터를 각 변수의 값으로 표현
values = list(df['사고건수 (건)'])
values += values[:1]
ax.plot(angles, values, linewidth=2, linestyle='solid', label="사고건수 (건)")
ax.fill(angles, values, alpha=0.4)

# 그래프 제목 설정
plt.title("행정구역별 사고건수 레이더 차트", size=14, color="blue", y=1.1)

# 레이더 차트 출력
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 주어진 데이터를 DataFrame으로 변환
data = {
    "행정구역별(1)": ["전국", "서울특별시", "서울특별시", "서울특별시", "서울특별시", "부산광역시", "부산광역시", "부산광역시"],
    "행정구역별(2)": ["소계", "소계", "종로구", "중구", "용산구", "소계", "중구", "서구"],
    "사고건수 (건)": [8889, 1026, 15, 21, 18, 443, 6, 14]
}

df = pd.DataFrame(data)

# Pivot을 통해 행정구역별(1), 행정구역별(2)를 인덱스로 가지는 데이터로 변환
heatmap_data = df.pivot(index="행정구역별(1)", columns="행정구역별(2)", values="사고건수 (건)")

# Seaborn을 사용하여 히트맵 그리기
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt="d", linewidths=.5)
plt.title("행정구역별 사고건수 히트맵")
plt.show()


```
